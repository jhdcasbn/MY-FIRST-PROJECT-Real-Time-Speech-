```python
# 1. Install the "Brain" tools
!pip install deep_translator -q

import re
from deep_translator import GoogleTranslator

# --- THE TASK ---

def clean_and_translate(raw_text):
    # A. DATA CLEANING (Regex)
    # This removes all the junk "Bulgaran", "Iphony", and symbols
    # Keeping only Hindi script (Unicode range \u0900-\u097F)
    cleaned_hindi = re.sub(r'[^\u0900-\u097F\s]', '', raw_text)
    cleaned_hindi = " ".join(cleaned_hindi.split()) # Fix double spaces

    if not cleaned_hindi:
        return "No Hindi detected to translate.", ""

    # B. TRANSLATION (The Brain)
    print("🧠 Processing translation...")
    translated_text = GoogleTranslator(source='hi', target='en').translate(cleaned_hindi)

    return cleaned_hindi, translated_text

# --- TEST IT WITH YOUR SONG TEXT ---
dirty_input = "माहारनि लागे मने Bulgaran C मीठे रस से भरयो री राधा रानी लागे Iphony"
hindi, english = clean_and_translate(dirty_input)

print("\n" + "="*50)
print(f"🇮🇳 CLEANED HINDI: {hindi}")
print("-" * 50)
print(f"🇺🇸 SMART ENGLISH: {english}")
print("="*50)
```

```python
# 1. Install the "Brain" tools
!pip install deep_translator -q
!pip install textblob -q

import time
from textblob import TextBlob
from deep_translator import GoogleTranslator

# --- THE TASK: Benchmarking Latency ---

def benchmark_pipeline(text):
    start_time = time.time()

    # 1. Translation Step
    translated = GoogleTranslator(source='hi', target='en').translate(text)

    # 2. Sentiment Step
    analysis = TextBlob(translated)

    end_time = time.time()
    total_latency = end_time - start_time

    return {
        "text_length": len(text),
        "latency": round(total_latency, 4),
        "sentiment": analysis.sentiment.polarity
    }

# --- EXECUTION ---
test_lyrics = "मीठे रस से भरयो री राधा रानी लागे"
stats = benchmark_pipeline(test_lyrics)

print("\n" + "📊" * 15)
print("WEEK 2 PERFORMANCE LOG")
print(f"Input Length: {stats['text_length']} characters")
print(f"Processing Time: {stats['latency']} seconds")
print(f"Confidence (Sentiment): {stats['sentiment']}")
print("📊" * 15)
```
